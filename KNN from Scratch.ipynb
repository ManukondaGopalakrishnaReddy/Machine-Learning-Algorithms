Data is already cleaned, so we can directly give it to the model. We can also do preprocessing, to churn out accuracy

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('Type',axis=1),df["Type"],test_size=0.2)
```

```python
# Extracting the values

x_train=x_train.values
x_test=x_test.values
y_train=y_train.values
y_test=y_test.values
```

```python
class KNN:
    def __init__(self,k):
        self.k=k
    
    # training the model
    def fit(self,x,y):
        self.x=x
        self.y=y
        
    # implementing eucledian distance
    def distance(self,x,x1):
        c=0
        for i in range(len(x)):
            c+=(x[i]-x1[i])**2
        return c**0.5
    
    # prediction on new values
    def predict(self,x1):
        l=[]
        for i in x1:
            d=[]
            for j in range(len(self.x)):
                p=self.distance(i,self.x[j])
                d.append([p,self.y[j]])
            d.sort(key=lambda p:p[0])
            di={}
            for j in range(self.k):
                di[d[j][1]]=di.get(d[j][1],0)+1
            p=list(di.items())
            p.sort(key=lambda a:a[1],reverse=True)
            l.append(p[0][0])
        return np.array(l)
```

```python
model=KNN(8)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# Accuracy of the custom coded model on the test-set

print('Accuracy is: ',accuracy_score(y_test,y_pred))
```

```python
# F1-Score of the custom coded model on the test-set

print('F1-Score is: ',f1_score(y_test,y_pred,average='weighted'))
```
